# **CERVICAL CANCER PREDICTION**

## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Understanding Data**

```python
df=pd.read_csv('/content/risk_factors_cervical_cancer.csv')
df
```

```python
df.info()
```

## **Data Cleaning**

```python
df.isnull().sum()
```

**No Null Value**

```python
df.replace('?',0,inplace=True)
df.head()
```

```python
sns.countplot(x='Smokes',hue='Biopsy',data=df)
```

```python
sns.boxplot(x='Hormonal Contraceptives',y='Age',data=df)
```

```python
df['Biopsy'].value_counts().plot.pie(colors=['paleturquoise','navy'],
                                     labels=['Not Present','Present'])
plt.show()
```

## **Correlational Analysis**

```python
cat_columns=df.select_dtypes(include='object').columns
for col in cat_columns:
  df[col] =pd.to_numeric(df[col])
  df[col]=df[col].astype(np.int64)
df.info()
```

```python
from imblearn.over_sampling import BorderlineSMOTE
```

```python
x=df.drop('Biopsy',axis=1)
y=df['Biopsy']
smote=BorderlineSMOTE(random_state=42)
X_resampled,y_resampled=smote.fit_resample(x,y)

df=pd.concat([X_resampled,y_resampled],axis=1)
```

```python
df['Biopsy'].value_counts()
```

```python
df.drop(['STDs:cervical condylomatosis','STDs:AIDS'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(df.corr(),annot=True,fmt='0.2f',cmap='magma')
```

**Dropping Columns with less correlation**

```python
columns_drop=['Biopsy',
              'STDs: Time since last diagnosis',
              'STDs: Time since first diagnosis',
              'STDs:HPV',
              'STDs',
              'Smokes (packs/year)',
              'STDs:HIV',
              'Dx:Cancer',
              'Dx:CIN',
              'Dx:HPV',
              'Dx',
              'STDs:Hepatitis B',
              'STDs:molluscum contagiosum',
              'STDs:pelvic inflammatory disease',
              'Hormonal Contraceptives',
              'Hormonal Contraceptives (years)',
              'IUD (years)',
              'STDs:condylomatosis',
              'STDs:genital herpes',
              'STDs:vulvo-perineal condylomatosis'
              ]
```

```python
X_resampled=df.drop(columns=columns_drop,axis=1)
X_resampled.info()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(xtrain,ytrain)
logistic.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

## **Data Scaling**

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
standard=StandardScaler()
x_stand=standard.fit_transform(X_resampled)
xtrains,xtests,ytrains,ytests=train_test_split(x_stand,y_resampled,test_size=0.2,random_state=42)
tree=DecisionTreeClassifier()
tree.fit(xtrains,ytrains)
tree.score(xtests,ytests)
```

```python
minmax=MinMaxScaler()
x_minmax=standard.fit_transform(X_resampled)
xtrainm,xtestm,ytrainm,ytestm=train_test_split(x_minmax,y_resampled,test_size=0.2,random_state=42)
tree=DecisionTreeClassifier()
tree.fit(xtrainm,ytrainm)
tree.score(xtestm,ytestm)
```

**We will take MinMaxScaler data**

## **Model Selection**

```python
log=LogisticRegression()
log.fit(xtrainm,ytrainm)
log.score(xtestm,ytestm)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(xtrainm,ytrainm)
svc.score(xtestm,ytestm)
```

```python
tree.fit(xtrainm,ytrainm)
tree.score(xtestm,ytestm)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier()
forest.fit(xtrainm,ytrainm)
forest.score(xtestm,ytestm)
```

```python
y_pred=forest.predict(xtestm)
```

## **Report**

```python
from sklearn.metrics import classification_report
report=classification_report(ytestm,y_pred)
print(report)
```
